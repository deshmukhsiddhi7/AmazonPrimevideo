```python
data.sort_values("show_id",ascending=True,inplace=False)
```

```python
sns.countplot(x='type',data=data,palette='plasma')
plt.xlabel('type',fontsize=20,alpha=0.7)
plt.ylabel('Count',fontsize=20,alpha=0.7)
plt.title('Comparing Movies and TV Show(Series)',fontsize=25)
plt.show()
```

```python
data['release_year'].unique()
```

```python
data.groupby('rating').mean()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(y=data['release_year'],color='#C9A0DC')
plt.yticks(rotation=20)
plt.title("Movie and Series Release Year Count Plot",fontsize=20)
plt.xlabel('Counts',fontsize=20,alpha=0.7)
plt.ylabel('Release Year',fontsize=20,alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Top 15 Countries on the basis of Content Creation ",fontsize=25)
sns.countplot(data=data,y="country",order=data['country'].value_counts().index[0:15],palette='coolwarm')
plt.xlabel('Count',fontsize=20,alpha=0.7)
plt.ylabel('Countries',fontsize=20,alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='release_year', y='rating',hue="type", data=data, color='skyblue')
plt.title("Scatter PLOT")
plt.xlabel("title")
plt.ylabel("duration")
plt.show()
```

```python
sns.displot(data['rating'],bins=40)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
sns.barplot(x='release_year',y="rating", hue='type', data=data, ci=None, palette='Set2')
plt.title("Distribution of rating with type of movie")
plt.xlabel("Release year")
plt.ylabel("Count")
plt.show()
```
